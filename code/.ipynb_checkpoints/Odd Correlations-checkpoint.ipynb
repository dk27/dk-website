```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
df = pd.read_csv("full_dataset.csv")
```

```python
# drop columns with more than 30% of missing values
df = df.dropna(thresh=len(df)*0.7, axis=1)
# drop columns where very value is the same
distinct = df.apply(pd.Series.nunique)
drop = distinct[distinct == 1].index
df = df.drop(drop, axis=1)
```

```python
list(df.columns)
```

```python
# create a line plot with 2 y axis
from matplotlib.ticker import MaxNLocator
def plot(series1_name, series2_name, x_name, r2):
    df1 = df[[series1_name, series2_name, x_name]]
    df1 = df1.dropna()
    fig, ax1 = plt.subplots()   
    
    ax1.plot(df1[x_name], df1[series1_name], color='lightcoral')
    ax1.set_xlabel(x_name, fontsize=14)
    ax1.set_ylabel(series1_name, color='lightcoral')
    ax1.xaxis.set_major_locator(MaxNLocator(integer=True))

    ax2 = ax1.twinx()
    ax2.plot(df1[x_name], df1[series2_name], color='deepskyblue')
    ax2.set_xlabel(x_name, fontsize=14)
    ax2.set_ylabel(series2_name, color='deepskyblue')
    ax2.xaxis.set_major_locator(MaxNLocator(integer=True))

    plt.title('{} vs {}\n Corr={}'.format(series1_name, series2_name, r2))
    plt.show()
    fig.savefig('{}.png'.format(series1_name+series2_name))
```

```python
corr = df.drop('Year', axis=1).corr()
```

```python
for i in range(80):
    col = corr.columns[i] 
    temp = corr[[col]].sort_values(by=col, ascending = False)
    idx = temp.iloc[1].name
    r2 = temp.iloc[1]
    plot(col, idx, 'Year', float(round(r2,3)))
```

```python
lst = [*range(24,31), *range(39,70)]
df_dairy_crime = df.iloc[:,lst].dropna()
```

```python
dairy_crime_corr = df_dairy_crime.corr()
```

```python
sn.set_theme(style="white")
f, ax = plt.subplots(figsize=(15,12))
mask = np.triu(np.ones_like(dairy_crime_corr, dtype=bool))
cmap = sn.diverging_palette(250,15, as_cmap=True)
hm = sn.heatmap(dairy_crime_corr, 
           cmap=cmap, 
           mask=mask, 
           square = True, 
           linewidth = 0.5, 
           cbar_kws = {"shrink": 0.5},
           center = 0
          )
figure = hm.get_figure()
figure.savefig('crime_dariy_corr.png', dpi=200)
```
